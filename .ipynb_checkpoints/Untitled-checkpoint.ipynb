```python
import re
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def code2mat(code):
    pattern = r"pt3dadd\((\d+?.\d+?,\d+?.\d+?,\d+?.\d+?\,\d+?.\d+?,\d)\)"
    code = re.findall(pattern, code)
    data = np.array([list(map(float, c.split(",")))[:4] for c in code])
    return data


def get_components(path):
    with open(path, 'r') as f:
        text = f.read()
    pattern = re.compile(r"filament_(\d+?)\[\d+?\] \{(.+?)\}", re.S)
    components = {}
    for idx, code in re.findall(pattern, text):
        if idx not in components:
            components[idx] = []
        components[idx].append(code2mat(code))
    return components


comp_dict = get_components('1.hoc')
tmp = []
for mats in comp_dict.values():
    tmp += mats

origin = np.array([d[0, :3] for d in tmp])
tail = np.array([d[-1, :3] for d in tmp])
```

```python
from queue import Queue

class FilamentNode(object):
    
    def __init__(self, coord):
        self.coordinate = coord
        self.neighboors = []
        self.length = []
        self.width = []
    
    def __hash__(self):
        return self.coordinate.__hash__()
    
    def __str__(self):
        return 'FilamentNode' + str(self.coordinate)
    
    __repr__ = __str__
    

class Filament(object):
    
    def __init__(self, root):
        self.root = root
        # dfs
        viewed = set([self.root])
        q = Queue()
        pair = (self.root, 0)
        q.put(pair)
        out_list = [pair]
        while not q.empty():
            node, depth = q.get()
            for tmp in node.neighboors:
                if tmp not in viewed:
                    viewed.add(tmp)
                    pair = (tmp, depth + 1)
                    q.put(pair)
                    out_list.append(pair)
    
    
def get_node(coord, d):
    if coord not in d:
        d[coord] = FilamentNode(coord)     
    return d[coord]
```

```python
d = {}
for curve in tmp:
    # nodes
    origin = get_node(tuple(curve[0, :3]), d)
    tail = get_node(tuple(curve[-1, :3]), d)
    # link nodes
    origin.neighboors.append(tail)
    tail.neighboors.append(origin)
    # average width
    width = curve[:, -1].mean()
    origin.width.append(width)
    tail.width.append(width)
    # curve length
    length = np.sqrt(np.square(np.diff(curve[:, :3], axis=0)).sum(axis=1)).sum()
    origin.length.append(length)
    tail.length.append(length)


root1 = (188.153 , 450.392 , 393.844)
root2 = (71.1441 , 166.232  , 674.555)
root3 = (3582.35  , 4552.36  , 1046.71)
root4 = (3567.71  ,  9267.96  ,  1440.6)
    
    
result = Filament(get_node(root2, d)).bfs()

coords = np.array([c[0].coordinate for c in result])
l = [c[1] for c in result]
```
